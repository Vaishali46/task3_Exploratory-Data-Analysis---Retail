### Data Visualization

```python
sns.pairplot(df)
```

```python
fig, ax = plt.subplots(1, 5, figsize=(15,3))

sns.boxplot(df["Postal Code"], ax=ax[0])
sns.boxplot(df['Sales'], ax=ax[1])
sns.boxplot(df['Quantity'], ax=ax[2])
sns.boxplot(df["Discount"] ,ax=ax[3])
sns.boxplot(df["Profit"], ax=ax[4])
```

Checking Correlation

```python
sns.heatmap(df.corr(), cmap="icefire", annot= True)
```

From plot :

    Sales and Profit have a direct relationship
    Sales and Discount have inverse relationship
    Profit and Discount have inverse relationship

```python
x, y = plt.subplots(nrows = 2, ncols = 2, figsize=(12,8))

sns.countplot(df["Ship Mode"], ax=y[0][0])
sns.countplot(df["Segment"], ax=y[0][1])
sns.countplot(df["Region"], ax=y[1][0])
sns.countplot(df["Category"], ax=y[1][1])
```

```python
x, y = plt.subplots(figsize=(18,6))
sns.countplot(df['Sub-Category'])
plt.title("Sub-Category")
```

```python
x, y = plt.subplots(figsize=(8,3))
sns.countplot(df['Category'])
plt.title("Category")
```

```python
x = df['State'].value_counts()
x = x.head(10)
x.plot(kind='bar', figsize=(10,5))
plt.xlabel("States")
plt.ylabel("Numbers")
plt.title("States with Top Sales")
```

```python
x = df['City'].value_counts()
x = x.head(10)
x.plot(kind='bar', figsize=(10,5))
plt.xlabel("City")
plt.ylabel("Numbers")
plt.title("Cities with Top Sales")
```

Profit, Discount and Sales Based on Ship mode

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))
sns.barplot(df["Ship Mode"], df["Profit"], ax=ax[0])
sns.barplot(df["Ship Mode"], df["Discount"], ax=ax[1])
sns.barplot(df["Ship Mode"], df["Sales"], ax=ax[2])
```
